```python
import pandas as pd
df= pd.read_csv('pizza_sales.csv')
df
```

```python
df['order_date'] = df['order_date'].str.replace('/', '-')
```

```python
df['order_date'] = pd.to_datetime(df['order_date'], format='%d-%m-%Y')
```

```python
PilihKolom = df[['order_id', 'order_date','quantity', 'pizza_name']]
PilihKolom
```

```python
from datetime import timedelta
# df_sorted_desc = PilihKolom.sort_values(by='order_date', ascending=False)

tanggal_indeks = pd.to_datetime('2015-01-01')
i = 0
tanggal_akhir = pd.to_datetime('2015-12-31')
while tanggal_indeks <= tanggal_akhir:
    tanggal_baru = tanggal_indeks + timedelta(days=5)
    
    # Gunakan .loc[] untuk assignment yang benar
    PilihKolom.loc[(PilihKolom['order_date'] >= tanggal_indeks) & 
                   (PilihKolom['order_date'] < tanggal_baru), 'cluster'] = i
    
    # Update tanggal_indeks dan increment nilai i
    tanggal_indeks = tanggal_baru
    i += 1
```

```python
# Kelompokkan berdasarkan minggu
grouped = PilihKolom.groupby('cluster')

# Pisahkan DataFrame untuk setiap minggu
weekly_dataframes = {week: group for week, group in grouped}

# Simpan setiap minggu ke file CSV (opsional) dan tampilkan hasil
for week, df in weekly_dataframes.items():
    print(f"Data untuk Minggu ke-{week}:")
    print(df)
    print("\n")

    # Opsional: Simpan ke file CSV
    # df.to_csv(f"week_{week}.csv", index=False)
```

```python
df_list = []
for i in range(len(weekly_dataframes)):
    # Menghitung jumlah quantity per pizza
    pizza_quantity = weekly_dataframes[i].groupby('pizza_name')['quantity'].sum()

    # Mengurutkan berdasarkan quantity dan mengambil yang teratas
    most_ordered_pizza = pizza_quantity.sort_values(ascending=False).head(3).reset_index()
    most_ordered_pizza['cluster'] = i

    df_list.append(most_ordered_pizza)

result_df = pd.concat(df_list, ignore_index=True)
result_df
```

```python
# Menggunakan pivot_table untuk membuat format one-hot encoding
one_hot = pd.get_dummies(result_df['pizza_name'])

# Menambahkan kolom Transaksi_ID
df_one_hot = result_df.join(one_hot)
df_one_hot
```

```python
df_one_hot.drop(['pizza_name', 'quantity'], axis=1, inplace=True)
```

```python
df_one_hot
```

```python
# Menggunakan pivot_table untuk menjadikan Transaksi_ID sebagai index dan barang sebagai kolom
df_pivot = df_one_hot.groupby(['cluster']).sum()
df_pivot
```

```python
df_pivot[df_pivot > 1] = 1
```

```python
df_pivot
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
# Menggunakan algoritma Apriori untuk menemukan itemset yang sering muncul
frequent_itemsets = apriori(df_pivot, min_support=0.15, use_colnames=True)
frequent_itemsets
```

```python
# Menghasilkan aturan asosiasi dari itemset yang sering muncul
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.1, num_itemsets=3)

# Menampilkan hasil aturan asosiasi
print("\nAturan Asosiasi yang ditemukan:")
rules
```

```python
# Fungsi untuk menghitung support
def calculate_support(itemset, transactions):
    count = 0
    for transaction in transactions:
        if all(item in transaction for item in itemset):
            count += 1
    return count / len(transactions)
```

```python
# Fungsi untuk menemukan frequent itemsets
def find_frequent_itemsets(transactions, min_support):
    itemsets = []
    items = {item for transaction in transactions for item in transaction}
    
    # Iterasi level itemsets
    k = 1
    while True:
        if k == 1:
            # Itemsets tunggal
            candidates = [{item} for item in items]
        else:
            # Gabungkan frequent itemsets sebelumnya
            candidates = [
                set(i).union(set(j)) for i in itemsets for j in itemsets
                if len(set(i).union(set(j))) == k
            ]
            # Hapus duplikat
            candidates = list(set(frozenset(c) for c in candidates))
        
        # Hitung support kandidat
        frequent_itemsets = []
        for candidate in candidates:
            support = calculate_support(candidate, transactions)
            if support >= min_support:
                frequent_itemsets.append(candidate)
                print(f"Frequent Itemset: {candidate}, Support: {support:.2f}")
        
        # Hentikan jika tidak ada frequent itemsets
        if not frequent_itemsets:
            break
        
        # Simpan frequent itemsets
        itemsets = frequent_itemsets
        k += 1
    
    return itemsets
```

```python
# Fungsi untuk menghasilkan aturan asosiasi
def generate_association_rules(frequent_itemsets, transactions, min_confidence):
    rules = []
    for itemset in frequent_itemsets:
        for i in range(1, len(itemset)):
            for antecedent in combinations(itemset, i):
                antecedent = set(antecedent)
                consequent = itemset - antecedent
                if consequent:
                    # Hitung confidence
                    support_itemset = calculate_support(itemset, transactions)
                    support_antecedent = calculate_support(antecedent, transactions)
                    confidence = support_itemset / support_antecedent
                    if confidence >= min_confidence:
                        rules.append((antecedent, consequent, confidence))
                        print(f"Rule: {antecedent} -> {consequent}, Confidence: {confidence:.2f}")
    return rules
```
