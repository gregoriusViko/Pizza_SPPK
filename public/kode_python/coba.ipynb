```python
# Importing the necessary libraries
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# Creating the dataset
dataset = [['Milk', 'Bread', 'Butter'],
           ['Bread', 'Eggs'],
           ['Milk', 'Eggs', 'Butter'],
           ['Milk', 'Bread', 'Eggs', 'Butter'],
           ['Milk', 'Bread', 'Eggs']]# Encoding the dataset
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)# Finding the frequent itemsets
frequent_itemsets = apriori(df, min_support=0.2, use_colnames=True)# Generating the association rules
frequent_itemsets
```
